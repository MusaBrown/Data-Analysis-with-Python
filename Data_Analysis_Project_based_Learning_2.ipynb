## Handle Missing Values

### Subtask:
Identify if there are any missing values in the dataset and propose strategies to handle them, such as imputation or removal, based on the data's characteristics.


**Reasoning**:
To address the subtask, I will first calculate the total and percentage of missing values for each column. Then, I will filter for columns that actually have missing values and display them sorted by the percentage of missing values in descending order.



```python
missing_values = df.isnull().sum()
missing_percentage = (df.isnull().sum() / len(df)) * 100

missing_info = pd.DataFrame({
    'Missing Count': missing_values,
    'Missing Percentage': missing_percentage
})

# Filter to show only columns with missing values and sort by percentage
missing_info = missing_info[missing_info['Missing Count'] > 0].sort_values(by='Missing Percentage', ascending=False)

print("Columns with Missing Values and their Percentages:")
display(missing_info)
```

### Strategies for Handling Missing Values

The previous analysis revealed that all missing values are concentrated in the 'Y' columns, which represent yearly data from 2001 to 2022. The percentages of missing values in these columns vary significantly, from approximately 5.9% to 83.0%.

Considering the nature of this time-series financial data, here are proposed strategies:

1.  **For columns with very high missing percentages (e.g., Y2001, Y2002, Y2003, Y2004, Y2005, Y2006, Y2007 with > 45% missing data):**
    *   **Recommendation**: **Consider dropping these columns.** With such a large proportion of missing data, any imputation method would likely introduce significant bias or inaccuracies, making the data in these columns unreliable for analysis. It's often better to work with fewer, more complete features than with many features that are heavily imputed.

2.  **For columns with moderate missing percentages (e.g., Y2008-Y2019 with 12.8% to 30.7% missing data):**
    *   **Recommendation**: **Advanced Imputation using Time-Series Methods.** Given that these are yearly financial data, a simple mean/median imputation might not be appropriate as it doesn't account for trends or seasonality. Instead, interpolation methods like `fillna(method='ffill')` (forward fill) or `fillna(method='bfill')` (backward fill) could be used. Linear or polynomial interpolation could also be considered if the data shows clear trends and the missing gaps are not too large. The choice depends on the underlying patterns of the financial data. For instance, if growth is generally steady, linear interpolation might be suitable.

3.  **For columns with low missing percentages (e.g., Y2020, Y2021, Y2022 with ~5.9% missing data):**
    *   **Recommendation**: **Targeted Imputation.** For these columns, imputation is more viable. Similar to the moderate percentage columns, time-series specific imputation methods like forward/backward fill or linear interpolation would be preferred over simple statistical measures to maintain the temporal integrity of the data. Alternatively, if these columns are critical, more sophisticated methods like K-Nearest Neighbors (KNN) imputation could be explored if other relevant columns are available to predict the missing values.

**Next Step**: Based on these considerations, we will proceed with implementing a selected strategy to handle the missing values in the 'Y' columns. For this exercise, we will apply **forward fill followed by backward fill** for the 'Y' columns to handle the moderate and low missing value percentages, and then re-evaluate the columns with high missing percentages.

**Reasoning**:
To implement the proposed strategy, I will first identify all 'Y' columns and then apply forward fill followed by backward fill to these columns to impute missing values. After imputation, I will check the remaining missing values.



```python
y_columns = [col for col in df.columns if col.startswith('Y')]

# Apply forward fill
df[y_columns] = df[y_columns].fillna(method='ffill')

# Apply backward fill
df[y_columns] = df[y_columns].fillna(method='bfill')

print("Missing values after forward and backward fill:")
missing_values_after_imputation = df[y_columns].isnull().sum()
missing_percentage_after_imputation = (df[y_columns].isnull().sum() / len(df)) * 100

missing_info_after_imputation = pd.DataFrame({
    'Missing Count': missing_values_after_imputation,
    'Missing Percentage': missing_percentage_after_imputation
})

# Filter to show only columns with remaining missing values and sort by percentage
missing_info_after_imputation = missing_info_after_imputation[missing_info_after_imputation['Missing Count'] > 0].sort_values(by='Missing Percentage', ascending=False)

if not missing_info_after_imputation.empty:
    print("Remaining missing values in 'Y' columns:")
    display(missing_info_after_imputation)
else:
    print("No missing values remaining in 'Y' columns.")
```

**Reasoning**:
The previous code produced `FutureWarning` messages due to the deprecated use of `DataFrame.fillna(method='ffill')` and `DataFrame.fillna(method='bfill')`. To address this, I will update the code to use the recommended `df.ffill()` and `df.bfill()` methods directly on the selected columns to ensure compliance with current pandas best practices.



```python
y_columns = [col for col in df.columns if col.startswith('Y')]

# Apply forward fill
df[y_columns] = df[y_columns].ffill()

# Apply backward fill
df[y_columns] = df[y_columns].bfill()

print("Missing values after forward and backward fill:")
missing_values_after_imputation = df[y_columns].isnull().sum()
missing_percentage_after_imputation = (df[y_columns].isnull().sum() / len(df)) * 100

missing_info_after_imputation = pd.DataFrame({
    'Missing Count': missing_values_after_imputation,
    'Missing Percentage': missing_percentage_after_imputation
})

# Filter to show only columns with remaining missing values and sort by percentage
missing_info_after_imputation = missing_info_after_imputation[missing_info_after_imputation['Missing Count'] > 0].sort_values(by='Missing Percentage', ascending=False)

if not missing_info_after_imputation.empty:
    print("Remaining missing values in 'Y' columns:")
    display(missing_info_after_imputation)
else:
    print("No missing values remaining in 'Y' columns.")
```

To fulfill the third instruction of the subtask, I will transform the DataFrame from a wide format to a long format. This involves using the `melt` function to convert the 'Y' columns (representing years) into a 'Year' column and their corresponding values into a 'Value' column. ## Perform Exploratory Data Analysis (EDA)

Calculate and display the total 'Value' for each year and identify the years with the highest and lowest total values.
